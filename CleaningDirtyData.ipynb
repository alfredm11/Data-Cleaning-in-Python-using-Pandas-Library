Let's import pandas library to clean this dirty file.

```python
import pandas as pd
```

Let's load our dataset and put it in df dataframe and have a quick look using head function.
This data was retrieved from: https://foresightbi.com.ng/microsoft-power-bi/dirty-data-samples-to-practice-on/

```python
df = pd.read_excel('2.-Badly-Structured-Sales-Data-2.xlsx')
df.head(5)
```

Let's look some more info about our dataset.

```python
df.info()
```

So our columns was arranged based on Ship Mode(First Class, Same Day, Second Class, Standard Class) and on the first row we can see the Segments(Consumer, Corporate, Home Office). So my goal is to create a separate dataframe for each Segment's ship mode. The values under the Ship Mode column are dates. So we will rename that column to Order_Date later. 

```python
df.head(4)
```

Let's create dataset for each Segment's ship Mode

```python
# #Legend:
# fc_df -- first class dataframe
# sd_df -- same day dataframe
# sc_df -- second class dataframe
# stdc_df -- standard class dataframe



fc_df = df[['Ship Mode', 'First Class', 'Unnamed: 2', 'Unnamed: 3']]
fc_df = fc_df.rename(columns={"Ship Mode": "Order_Date",
              "First Class": "Consumer",
              "Unnamed: 2": "Corporate",
              "Unnamed: 3": "Home_Office"})


sd_df = df[['Ship Mode', 'Same Day', 'Unnamed: 5', 'Unnamed: 6']]
sd_df = sd_df.rename(columns={"Ship Mode": "Order_Date",
              "Same Day": "Consumer",
              "Unnamed: 5": "Corporate",
              "Unnamed: 6": "Home_Office"})

sc_df = df[['Ship Mode', 'Second Class', 'Unnamed: 8', 'Unnamed: 9']]
sc_df = sc_df.rename(columns={"Ship Mode": "Order_Date",
              "Second Class": "Consumer",
              "Unnamed: 8": "Corporate",
              "Unnamed: 9": "Home_Office"})

stdc_df = df[['Ship Mode', 'Standard Class', 'Unnamed: 11', 'Unnamed: 12']]
stdc_df = stdc_df.rename(columns={"Ship Mode": "Order_Date",
              "Standard Class": "Consumer",
              "Unnamed: 11": "Corporate",
              "Unnamed: 12": "Home_Office"})
```

Let's view the first three rows of our dataframe for our reference.

```python
df.head(3)
```

After creating the dataframes, let's remove the first two rows(row 0 and 1) of each dataframe because as you can see they won't add value to our newly created dataframes. We only used the first row as a reference for the segment category. If you can see the pattern, the columns two to three are for the First Class ship mode and each column is categorize for each segment. So the second column (First Class) is for the Consumer first class shipment, the third column(Unnamed:2) is for Corporate first class ship mode, the fourth column (Unnamed: 3) is for Home Office first class ship mode and so on.

```python
fc_df =  fc_df.iloc[2:]
sd_df =  sd_df.iloc[2:]
sc_df =  sc_df.iloc[2:]
stdc_df =  stdc_df.iloc[2:]
```

Then let's convert the dtypes of each colummn and check.

```python
fc_df = fc_df.convert_dtypes()
sd_df = sd_df.convert_dtypes()
sc_df = sc_df.convert_dtypes()
stdc_df = stdc_df.convert_dtypes()
```

```python
fc_df.info()
```

And now let's create a new dataset for each segment.

```python
fc_con = fc_df[['Order_Date', 'Consumer']]
fc_corp = fc_df[['Order_Date', 'Corporate']]
fc_ho = fc_df[['Order_Date', 'Home_Office']]

sd_con = sd_df[['Order_Date', 'Consumer']]
sd_corp = sd_df[['Order_Date', 'Corporate']]
sd_ho = sd_df[['Order_Date', 'Home_Office']]

sc_con =sc_df[['Order_Date', 'Consumer']]
sc_corp = sc_df[['Order_Date', 'Corporate']]
sc_ho = sc_df[['Order_Date', 'Home_Office']]

stdc_con = stdc_df[['Order_Date', 'Consumer']]
stdc_corp = stdc_df[['Order_Date', 'Corporate']]
stdc_ho = stdc_df[['Order_Date', 'Home_Office']]
```

Let's view our new dataframes

```python
fc_con.head(5)
```

```python
sd_corp.head(5)
```

```python
sc_ho.head(5)
```

Let's now rename our second column to Sales.

```python
fc_con = fc_con.rename(columns={"Consumer":"Sales"})
fc_corp = fc_corp.rename(columns={"Corporate":"Sales"})
fc_ho = fc_ho.rename(columns={"Home_Office":"Sales"})

sd_con = sd_con.rename(columns={"Consumer":"Sales"})
sd_corp = sc_corp.rename(columns={"Corporate":"Sales"})
sd_ho = sd_ho.rename(columns={"Home_Office":"Sales"})

sc_con = sc_con.rename(columns={"Consumer":"Sales"})
sc_corp = sc_corp.rename(columns={"Corporate":"Sales"})
sc_ho = sc_ho.rename(columns={"Home_Office":"Sales"})

stdc_con = stdc_con.rename(columns={"Consumer":"Sales"})
stdc_corp = stdc_corp.rename(columns={"Corporate":"Sales"})
stdc_ho = stdc_ho.rename(columns={"Home_Office":"Sales"})

```

Let's check our dataframe again. So we have two columns. Later we will add two columns for the Ship Mode and Segment.

```python
fc_con.head(5)
```

Let's now drop our NAs. In practice, you don't just remove NaNs/NAs because you just like to remove it. You must ask your stakeholders first before removing/dropping anything. But if you own the data, you can do whatever you want.

```python
fc_con = fc_con.dropna()
fc_corp = fc_corp.dropna()
fc_ho = fc_ho.dropna()

sd_con = sd_con.dropna()
sd_corp = sd_corp.dropna()
sd_ho = sd_ho.dropna()

sc_con = sd_con.dropna()
sc_corp = sd_corp.dropna()
sc_ho = sd_ho.dropna()

stdc_con = stdc_con.dropna()
stdc_corp = stdc_corp.dropna()
stdc_ho = stdc_ho.dropna()
```

Now let's check again our dataframe.

```python
fc_con.head(5)
```

Let's now add the two columns and fill it with string values.

```python
fc_con['Ship_Mode'] = "First Class"
fc_con['Segment'] = "Consumer"

fc_corp['Ship_Mode'] = "First Class"
fc_corp['Segment'] = "Corporate"

fc_ho['Ship_Mode'] = "First Class"
fc_ho['Segment'] = "Home_Office"

sd_con['Ship_Mode'] = "Same Day"
sd_con['Segment'] = "Consumer"

sd_corp['Ship_Mode'] = "Same Day"
sd_corp['Segment'] = "Corporate"

sd_ho['Ship_Mode'] = "Same Day"
sd_ho['Segment'] = "Home_Office"

sc_con['Ship_Mode'] = "Second Class"
sc_con['Segment'] = "Consumer"

sc_corp['Ship_Mode'] = "Second Class"
sc_corp['Segment'] = "Corporate"

sc_ho['Ship_Mode'] = "Second Class"
sc_ho['Segment'] = "Home_Office"

stdc_con['Ship_Mode'] = "Standard Class"
stdc_con['Segment'] = "Consumer"

stdc_corp['Ship_Mode'] = "Standard Class"
stdc_corp['Segment'] = "Corporate"

stdc_ho['Ship_Mode'] = "Standard Class"
stdc_ho['Segment'] = "Home_Office"
```

```python
stdc_ho.head(5)
```

Let's now combine our dataframes using concat function.

```python
fc_datasets = [fc_con, fc_corp, fc_ho]
fc_cleaned = pd.concat(fc_datasets)

sd_datasets = [sd_con, sd_corp, sd_ho]
sd_cleaned = pd.concat(sd_datasets)

sc_datasets = [sc_con, sc_corp, sc_ho]
sc_cleaned = pd.concat(sc_datasets)

stdc_datasets = [stdc_con, stdc_corp, stdc_ho]
stdc_cleaned = pd.concat(stdc_datasets)


```

Let's check our dataframes.

```python
fc_cleaned.head(5)
```

```python
sd_cleaned.head(5)
```

```python
sc_cleaned.head(5)
```

```python
stdc_cleaned.head(5)
```

Let's now finally combine all of our cleaned datasets into one. 

```python
all_dfs = [fc_cleaned, sd_cleaned, sc_cleaned, stdc_cleaned]
cleaned_df = pd.concat(all_dfs)
```

Let's check it.

```python
cleaned_df
```

And now, let's rearrange our columns to Ship Mode, Segment, Order_Date, and Sales.

```python
cleaned_df = cleaned_df[['Ship_Mode', 'Segment', 'Order_Date', 'Sales']]
cleaned_df
```

And for the last part, as you can see the indexes are not looking great. We will fix that using numpy.

```python
import numpy as np
```

```python
cleaned_df.index = np.arange(1, len(cleaned_df) + 1)
cleaned_df
```

Everything looks great. And now, let's export our dataframe to excel. 

```python
cleaned_df.to_excel("cleaned_data2.xlsx", sheet_name="Clean")
```
